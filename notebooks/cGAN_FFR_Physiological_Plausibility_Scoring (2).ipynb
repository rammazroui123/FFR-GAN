```python
# 1) Imports
import os, warnings
warnings.filterwarnings("ignore")  # hide warning messages to keep output cleaner

import numpy as np                  # numerical operations / arrays
import pandas as pd                 # table handling / CSV reading
import matplotlib.pyplot as plt     # plotting
import tensorflow as tf             # deep learning framework
from tensorflow.keras import layers, Model
from sklearn.model_selection import train_test_split   # split data into train/val/test
from sklearn.preprocessing import MinMaxScaler         # rescale features to [0,1]

print("TensorFlow version:", tf.__version__)
np.random.seed(42)                  # fix NumPy randomness for reproducibility
tf.random.set_seed(42)              # fix TensorFlow randomness for reproducibility


# 2) File paths
REAL_FFR_PATH    = "/content/real_FFR_summary.csv"          # trusted FFR file
FAKE_FFR_PATH    = "/content/fake_FFR_summary.csv"          # biased FFR file
REAL_DESIGN_PATH = "/content/real_lhs_stenosis_design.csv"  # eference geometry/design file
FAKE_DESIGN_PATH = "/content/fake_lhs_stenosis_design.csv"  # biased geometry/design file


# 3) Load CSV files
real_ffr    = pd.read_csv(REAL_FFR_PATH)       # load reference FFR values
fake_ffr    = pd.read_csv(FAKE_FFR_PATH)       # load biased FFR values
real_design = pd.read_csv(REAL_DESIGN_PATH)    # load reference anatomy/design features
fake_design = pd.read_csv(FAKE_DESIGN_PATH)    # load biased anatomy/design features

for df in [real_ffr, fake_ffr, real_design, fake_design]:
    df.columns = [c.strip() for c in df.columns]   # remove extra spaces from column names


# 4) Create a shared merge key and merge files
# Extract the shared middle identifier from each case name / stenosis ID
# so the FFR file and the design file can be matched case-by-case.

real_ffr['key']    = real_ffr['case_name'].str.split('_').str[1]
fake_ffr['key']    = fake_ffr['case_name'].str.split('_').str[1]
real_design['key'] = real_design['stenosis_ID'].str.split('_').str[1]
fake_design['key'] = fake_design['stenosis_ID'].str.split('_').str[1]

# Inner join = keep only rows where the same key exists in both files
real_df = real_ffr.merge(real_design, on='key', how='inner')
fake_df = fake_ffr.merge(fake_design, on='key', how='inner')


# 5) Clean raw columns + engineer extra features
numeric_cols = [
    "FFR", "artery_length_mm", "diameter_mm", "stenosis_position_mm",
    "stenosis_length_mm", "stenosis_severity_pct", "start_mm", "end_mm"
]

for df in [real_df, fake_df]:
    # convert important columns to numeric; invalid strings become NaN
    for c in numeric_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # convert eccentricity into a numeric binary variable
    # TRUE -> 1, FALSE -> 0
    df["is_eccentric"] = (
        df["is_eccentric"]
        .map({"TRUE": 1, "FALSE": 0, True: 1, False: 0})
        .fillna(0)
        .astype(float)
    )

    # convert stenosis severity percentage into a fraction if needed
    df["severity_frac"] = np.where(
        df["stenosis_severity_pct"] > 1,
        df["stenosis_severity_pct"] / 100,
        df["stenosis_severity_pct"]
    )

    # estimated diameter reduction in mm
    df["stenosis_diameter_mm"] = df["diameter_mm"] * df["severity_frac"]

    # compute lesion length from end minus start location
    df["lesion_length_mm"] = df["end_mm"] - df["start_mm"]

    # simple heuristic expected FFR from severity
    # this is not an exact physiological equation, just a helpful engineered feature
    df["expected_ffr"] = 1.0 - df["severity_frac"] * 0.6

    # difference between observed FFR and heuristic expected FFR
    df["ffr_residual"] = df["FFR"] - df["expected_ffr"]

    # interaction feature: lets the model see FFR × severity jointly
    df["ffr_x_severity"] = df["FFR"] * df["severity_frac"]


# 6) Remove invalid FFR values
# FFR is treated as physiologically meaningful only in [0,1] here.
real_df = real_df[(real_df["FFR"] >= 0) & (real_df["FFR"] <= 1)].copy().reset_index(drop=True)
fake_df = fake_df[(fake_df["FFR"] >= 0) & (fake_df["FFR"] <= 1)].copy().reset_index(drop=True)

print(f"Real: {len(real_df)}  Fake: {len(fake_df)}")


# 7) Select final modeling features
condition_features = [
    "diameter_mm",
    "stenosis_severity_pct",
    "severity_frac",
    "stenosis_diameter_mm",
    "lesion_length_mm",
    "is_eccentric",
    "expected_ffr",
    "ffr_residual",
    "ffr_x_severity"
]

# first column is the FFR value being judged, remaining columns are the conditions
feature_cols = ["FFR"] + condition_features

# keep only case name + final features, and drop rows with missing values
real_model = real_df[["case_name"] + feature_cols].dropna().reset_index(drop=True)
fake_model = fake_df[["case_name"] + feature_cols].dropna().reset_index(drop=True)


# 8) Min-Max scale features to [0,1]
# Scaler is fit on the reference data only and then applied unchanged to the fake data
scaler = MinMaxScaler()
real_scaled_arr = np.clip(scaler.fit_transform(real_model[feature_cols]), 0, 1)
fake_scaled_arr = np.clip(scaler.transform(fake_model[feature_cols]),     0, 1)

# turn scaled arrays back into labeled DataFrames - gives columns names back
real_scaled = pd.DataFrame(real_scaled_arr, columns=feature_cols)
real_scaled.insert(0, "case_name", real_model["case_name"].values)

fake_scaled = pd.DataFrame(fake_scaled_arr, columns=feature_cols)
fake_scaled.insert(0, "case_name", fake_model["case_name"].values)


# 9) Split only the REAL dataset
# 70% train, 15% val, 15% test
# Fake stays completely separate for evaluation only.
real_train_df, real_temp_df = train_test_split(real_scaled, test_size=0.30, random_state=42)
real_val_df,   real_test_df = train_test_split(real_temp_df, test_size=0.50, random_state=42)

# convert to NumPy arrays for TensorFlow - for cGAN training
train_array = real_train_df[feature_cols].values.astype("float32")
val_array   = real_val_df[feature_cols].values.astype("float32")
test_array  = real_test_df[feature_cols].values.astype("float32")
fake_array  = fake_scaled[feature_cols].values.astype("float32")

# define sizes input generator/discriminator will use
ffr_dim   = 1                          # one FFR input
cond_dim  = len(condition_features)    # 9 condition features
noise_dim = 16                         # random generator noise length

print(f"Train: {len(train_array)}  Val: {len(val_array)}  Test: {len(test_array)}  Fake: {len(fake_array)}")
print(f"cond_dim: {cond_dim}  noise_dim: {noise_dim}")

# 10) Build the generator
# Generator takes:
#   - random noise
#   - condition features
# and outputs:
#   - one generated FFR value in [0,1]

def build_generator(noise_dim, cond_dim):
    noise_in = layers.Input(shape=(noise_dim,), name="noise_input") # - input 16 values
    cond_in  = layers.Input(shape=(cond_dim,),  name="cond_input") # - input 9 values

    # combine noise and conditions - creates an FFR based on anatomy
    x = layers.Concatenate()([noise_in, cond_in])

    # dense network to transform noise+conditions into a candidate FFR
    x = layers.Dense(128)(x) # 128 nuerons
    x = layers.LeakyReLU(0.2)(x) # activation f(x) that introduces non-linearity, keeps 20% of - values instead of making them 0, which helps network learn more smoothly
    x = layers.BatchNormalization()(x) # normalizes activations inside the network

    # same size block for network to have more depth, instead of a single mapping
    x = layers.Dense(128)(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(64)(x) # compresses learned info from 128 to 64 units
    x = layers.LeakyReLU(0.2)(x)

    # sigmoid output ensures generated 1 FFR value in [0,1]
    out = layers.Dense(1, activation="sigmoid", name="generated_ffr")(x)

    # Generator takes noise + condition features as input and outputs one generated FFR value
    return Model([noise_in, cond_in], out, name="Generator")


# 11) Build the discriminator
# Discriminator takes:
#   - one FFR value
#   - the 9 condition features
# and outputs:
#   - a plausibility score in [0,1]

def build_discriminator(cond_dim):
    ffr_in  = layers.Input(shape=(1,),        name="ffr_input")
    cond_in = layers.Input(shape=(cond_dim,), name="cond_input")

    # combine FFR and conditions
    x = layers.Concatenate()([ffr_in, cond_in])

    # larger network because discriminator is the key final model
    x = layers.Dense(256)(x) # 256 dense neurons
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.3)(x) # 30% dropout rate to reduce overfitting

    x = layers.Dense(128)(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Dense(64)(x)
    x = layers.LeakyReLU(0.2)(x)

    x = layers.Dense(32)(x) # compresses final decision
    x = layers.LeakyReLU(0.2)(x)

    # final plausibility score
    out = layers.Dense(1, activation="sigmoid", name="physiological_score")(x)

    return Model([ffr_in, cond_in], out, name="Discriminator")


# create both models
generator     = build_generator(noise_dim, cond_dim)
discriminator = build_discriminator(cond_dim)

# show the model layers, shapes, and parameter counts
generator.summary()
discriminator.summary()


# 12) Unsupervised warm-up for the discriminator
# This stage trains the discriminator BEFORE full GAN training.
# It uses only real/reference data.
# Positives:
#   - correct (FFR, anatomy) pairs from real data
# Negatives:
#   - shuffled FFR with wrong anatomy
#   - correct FFR with shuffled anatomy
# This helps the discriminator learn "conditional consistency" before it ever sees generator outputs


print("\nUnsupervised warm-up")

# binary cross-entropy for plausible vs implausible pair classification
bce = tf.keras.losses.BinaryCrossentropy()

# mean squared error can be used as an additional stability / matching loss
mse = tf.keras.losses.MeanSquaredError()

# Adam optimizer updates discriminator weights during warm-up training by learning rate 0.001
warmup_optimizer = tf.keras.optimizers.Adam(0.001)

# makes TensorFlow dataset from real training data
warmup_ds = (
    tf.data.Dataset.from_tensor_slices(train_array)
    .shuffle(1000)
    .batch(64)
)

for epoch in range(80):

  # separate each batch into ffr value and its conditions
    for real_batch in warmup_ds:
        real_ffr_b  = real_batch[:, 0:1]   # real FFR values
        real_cond_b = real_batch[:, 1:]    # real conditions

        # Negative type 1:
        # shuffle FFR values within batch so they no longer match the anatomy
        idx_ffr      = tf.random.shuffle(tf.range(tf.shape(real_ffr_b)[0]))
        shuffled_ffr = tf.gather(real_ffr_b, idx_ffr)

        # Negative type 2:
        # shuffle anatomy conditions within batch so they no longer match the FFR
        idx_cond      = tf.random.shuffle(tf.range(tf.shape(real_cond_b)[0]))
        shuffled_cond = tf.gather(real_cond_b, idx_cond)

        # TensorFlow records operations so gradients can be computed and weights are updated
        with tf.GradientTape() as tape:
            # score matched real pairs - sends correct pairs to discriminator
            real_scores = discriminator([real_ffr_b, real_cond_b], training=True)

            # score mismatched pairs
            shuffled_ffr_sc  = discriminator([shuffled_ffr, real_cond_b],  training=True)
            shuffled_cond_sc = discriminator([real_ffr_b, shuffled_cond], training=True)

            # BCE:
            #   real pairs -> target near 1
            #   shuffled pairs -> target near 0

            # MSE term:
            #   explicitly pull real scores closer to 1.0
            loss = (
                bce(tf.ones_like(real_scores)       * 0.95, real_scores)      +                   # real matched pairs should be near 1
                bce(tf.zeros_like(shuffled_ffr_sc)  + 0.05, shuffled_ffr_sc)  +                 # shuffled FFR pairs should be near 0
                bce(tf.zeros_like(shuffled_cond_sc) + 0.05, shuffled_cond_sc) +          # shuffled anatomy pairs should be near 0
                mse(tf.ones_like(real_scores), real_scores) * 0.5                                  # make real scores near 1
            ) / 3.0

        grads = tape.gradient(loss, discriminator.trainable_variables)  # adjust weights
        warmup_optimizer.apply_gradients(zip(grads, discriminator.trainable_variables)) # update discriminator


# evaluate discriminator after warm-up
warmup_real = float(tf.reduce_mean(
    discriminator([train_array[:, 0:1], train_array[:, 1:]], training=False)
))

warmup_shuf_idx  = np.random.permutation(len(train_array))
warmup_shuf_ffr  = train_array[warmup_shuf_idx, 0:1]
warmup_shuf_cond = train_array[:, 1:]

warmup_shuf = float(tf.reduce_mean(
    discriminator([warmup_shuf_ffr, warmup_shuf_cond], training=False)
))

warmup_fake = float(tf.reduce_mean(
    discriminator([fake_array[:, 0:1], fake_array[:, 1:]], training=False)
))

print(f"After warm-up — D(real): {warmup_real:.3f}  D(shuffled): {warmup_shuf:.3f}  D(fake): {warmup_fake:.3f}")


# 13) Adversarial training settings
gen_optimizer  = tf.keras.optimizers.Adam(learning_rate=2e-4, beta_1=0.5) # generator weights will be updated
disc_optimizer = tf.keras.optimizers.Adam(learning_rate=8e-5, beta_1=0.5) # discriminator weights will be updated

DISC_UPDATES        = 2       # train discriminator twice per generator update
INPUT_NOISE_STD     = 0.003   # tiny noise for regularization
REAL_SCORE_TARGET   = 0.95    # desired real-score level
REAL_PENALTY_WEIGHT = 0.4     # extra penalty to keep real scores high


# 14) One discriminator training step
@tf.function
def discriminator_step(real_batch):
    batch_size  = tf.shape(real_batch)[0]
    real_ffr_b  = real_batch[:, 0:1]
    real_cond_b = real_batch[:, 1:]

    # generator creates fake FFR values conditioned on real anatomy
    noise   = tf.random.normal([batch_size, noise_dim])
    gen_ffr = generator([noise, real_cond_b], training=False)

    # also create shuffled hard negatives from the real batch
    idx          = tf.random.shuffle(tf.range(batch_size))
    shuffled_ffr = tf.gather(real_ffr_b, idx)

    # helper: add a little Gaussian noise, then clip back to [0,1]
    n = lambda t: tf.clip_by_value(
        t + tf.random.normal(tf.shape(t), stddev=INPUT_NOISE_STD), 0, 1
    )

    with tf.GradientTape() as tape:
        # discriminator scores:
        #   real matched pairs
        #   generator-made pairs
        #   shuffled mismatched pairs
        real_scores     = discriminator([n(real_ffr_b),   n(real_cond_b)], training=True)
        gen_scores      = discriminator([n(gen_ffr),      n(real_cond_b)], training=True)
        shuffled_scores = discriminator([n(shuffled_ffr), n(real_cond_b)], training=True)

        # BCE adversarial loss:
        #   real -> near 1
        #   fake / shuffled -> near 0
        adv_loss = (
            bce(tf.ones_like(real_scores)     * 0.95, real_scores)     +
            bce(tf.zeros_like(gen_scores)      + 0.05, gen_scores)      +
            bce(tf.zeros_like(shuffled_scores) + 0.05, shuffled_scores)
        ) / 3.0

        # extra penalty if real scores drop below target
        real_penalty = tf.reduce_mean(
            tf.square(tf.maximum(0.0, REAL_SCORE_TARGET - real_scores))
        )

        disc_loss = adv_loss + REAL_PENALTY_WEIGHT * real_penalty

    grads = tape.gradient(disc_loss, discriminator.trainable_variables)
    disc_optimizer.apply_gradients(zip(grads, discriminator.trainable_variables))
    return disc_loss


# 15) One generator training step
@tf.function
def generator_step(real_batch):
    batch_size  = tf.shape(real_batch)[0]
    real_cond_b = real_batch[:, 1:]
    noise       = tf.random.normal([batch_size, noise_dim])

    with tf.GradientTape() as tape:
        # generator proposes an FFR value from noise + real conditions
        gen_ffr = generator([noise, real_cond_b], training=True)

        # discriminator scores the generated pair
        gen_scores = discriminator([gen_ffr, real_cond_b], training=False)

        # generator wants discriminator to believe generated pairs are real/plausible
        loss = bce(tf.ones_like(gen_scores) * 0.95, gen_scores)

    grads = tape.gradient(loss, generator.trainable_variables)
    gen_optimizer.apply_gradients(zip(grads, generator.trainable_variables))
    return loss


# 16) Main adversarial training loop
EPOCHS     = 400
BATCH_SIZE = 32

train_dataset = (
    tf.data.Dataset.from_tensor_slices(train_array)
    .shuffle(1000, seed=42)
    .batch(BATCH_SIZE)
)

disc_losses, gen_losses      = [], []
val_real_hist, val_fake_hist = [], []

for epoch in range(EPOCHS):
    epoch_disc, epoch_gen = [], []

    for real_batch in train_dataset:
        # update discriminator multiple times per batch
        for _ in range(DISC_UPDATES):
            d_loss = discriminator_step(real_batch)

        # then update generator once
        g_loss = generator_step(real_batch)

        epoch_disc.append(float(d_loss))
        epoch_gen.append(float(g_loss))

    # store average losses for plotting
    disc_losses.append(np.mean(epoch_disc))
    gen_losses.append(np.mean(epoch_gen))

    # track average discriminator score on validation real data and on fake data
    v_real = float(tf.reduce_mean(
        discriminator([val_array[:, 0:1], val_array[:, 1:]], training=False)
    ))
    v_fake = float(tf.reduce_mean(
        discriminator([fake_array[:, 0:1], fake_array[:, 1:]], training=False)
    ))
    val_real_hist.append(v_real)
    val_fake_hist.append(v_fake)

    if (epoch + 1) % 50 == 0 or epoch == 0:
        print(
            f"Epoch {epoch+1:03d}/{EPOCHS} | D: {disc_losses[-1]:.4f} | "
            f"G: {gen_losses[-1]:.4f} | "
            f"D(real_val): {v_real:.3f} | D(fake): {v_fake:.3f}"
        )


# 17) Plot training curves
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("cGAN training curves", fontsize=13, fontweight="bold")

# left plot: discriminator and generator loss over time
ax1.plot(disc_losses, label="Discriminator loss", color="#185FA5")
ax1.plot(gen_losses,  label="Generator loss",     color="#993C1D")
ax1.axhline(0.693, color="gray", linestyle="--", alpha=0.5, label="Nash eq. (ln2)")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.legend()
ax1.grid(alpha=0.3)
ax1.spines[["top", "right"]].set_visible(False)

# right plot: mean discriminator score on validation real vs fake data
ax2.plot(val_real_hist, label="D(real val) → should → 1", color="#1D9E75")
ax2.plot(val_fake_hist, label="D(fake)     → should → 0", color="#EF9F27")
ax2.axhline(0.5, color="gray", linestyle="--", alpha=0.5, label="0.5 boundary")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Mean discriminator score")
ax2.set_ylim(0, 1.05)
ax2.legend()
ax2.grid(alpha=0.3)
ax2.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("/content/cgan_training_curves.png", dpi=150, bbox_inches="tight")
plt.show()


# 18) Final scoring function
# At the end, the discriminator is the final model.
# It takes:
#   - one FFR value
#   - 9 condition features
# and returns a plausibility score.
def get_scores(arr):
    ffr  = np.clip(arr[:, 0:1].astype("float32"), 0, 1)
    cond = np.clip(arr[:, 1:].astype("float32"),  0, 1)
    return discriminator.predict([ffr, cond], verbose=0).flatten()


# score each dataset split
real_train_scores = get_scores(train_array)
real_val_scores   = get_scores(val_array)
real_test_scores  = get_scores(test_array)
fake_scores       = get_scores(fake_array)


# 19) Print score summary
print("\nScore summary:")
for name, scores in [
    ("real_train", real_train_scores),
    ("real_val",   real_val_scores),
    ("real_test",  real_test_scores),
    ("fake_biased", fake_scores)
]:
    print(
        f"{name:20s} | n={len(scores):4d} | mean={np.mean(scores):.4f} | "
        f"std={np.std(scores):.4f} | min={np.min(scores):.4f} | max={np.max(scores):.4f}"
    )


# 20) Percentile-based threshold analysis
# These thresholds come from the real test score distribution.
# Then we check how many fake samples fall below them.
for pct in [10, 25]:
    thr = np.percentile(real_test_scores, pct)
    r   = np.mean(real_test_scores < thr)
    f   = np.mean(fake_scores < thr)

    print(f"\n{pct}th percentile threshold: {thr:.4f}")
    print(f"  Real test below: {r*100:.1f}%  |  Fake below: {f*100:.1f}%")


# 21) Plot score distributions
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Physiological plausibility scores", fontsize=13, fontweight="bold")

# histogram comparison
axes[0].hist(real_test_scores, bins=30, alpha=0.65, label="Real test",
             color="#1D9E75", edgecolor="white", density=True)
axes[0].hist(fake_scores, bins=30, alpha=0.65, label="Fake",
             color="#EF9F27", edgecolor="white", density=True)
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Density")
axes[0].set_title("Score distributions")
axes[0].legend()
axes[0].grid(alpha=0.3)
axes[0].spines[["top", "right"]].set_visible(False)

# cumulative distributions
axes[1].plot(np.sort(real_test_scores), np.linspace(0, 1, len(real_test_scores)),
             label="Real test", color="#1D9E75", linewidth=2)
axes[1].plot(np.sort(fake_scores), np.linspace(0, 1, len(fake_scores)),
             label="Fake", color="#EF9F27", linewidth=2)
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Cumulative proportion")
axes[1].set_title("Cumulative distributions")
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].spines[["top", "right"]].set_visible(False)

# score vs raw FFR values
real_test_ffr = real_test_df["FFR"].values
fake_ffr_vals = fake_model["FFR"].values

sc = axes[2].scatter(real_test_ffr, real_test_scores, c=real_test_scores,
                     cmap="RdYlGn", alpha=0.5, s=15, vmin=0, vmax=1, label="Real test")
axes[2].scatter(fake_ffr_vals, fake_scores, c=fake_scores,
                cmap="RdYlGn", alpha=0.2, s=10, vmin=0, vmax=1, marker="x", label="Fake")
axes[2].axvline(0.8, color="gray", linestyle=":", linewidth=1, label="Clinical FFR=0.8")
axes[2].set_xlabel("FFR value")
axes[2].set_ylabel("Plausibility score")
axes[2].set_title("Score vs FFR")
axes[2].legend(fontsize=8)
axes[2].grid(alpha=0.3)
axes[2].spines[["top", "right"]].set_visible(False)
plt.colorbar(sc, ax=axes[2], label="Score")

plt.tight_layout()
plt.savefig("/content/score_distributions.png", dpi=150, bbox_inches="tight")
plt.show()


# 22) Save per-case scores
all_scores = pd.concat([
    real_train_df[["case_name"]].assign(dataset="real_train", score=real_train_scores),
    real_val_df[["case_name"]].assign(dataset="real_val",     score=real_val_scores),
    real_test_df[["case_name"]].assign(dataset="real_test",   score=real_test_scores),
    fake_scaled[["case_name"]].assign(dataset="fake_biased",  score=fake_scores),
], ignore_index=True)

all_scores.to_csv("/content/physiological_scores_results.csv", index=False)



# 23) Save summary table
summary_table = pd.DataFrame({
    "dataset":    ["real_train", "real_val", "real_test", "fake_biased"],
    "n_samples":  [len(real_train_scores), len(real_val_scores),
                   len(real_test_scores),  len(fake_scores)],
    "mean_score": [np.mean(real_train_scores), np.mean(real_val_scores),
                   np.mean(real_test_scores),  np.mean(fake_scores)],
    "std_score":  [np.std(real_train_scores),  np.std(real_val_scores),
                   np.std(real_test_scores),   np.std(fake_scores)],
    "min_score":  [np.min(real_train_scores),  np.min(real_val_scores),
                   np.min(real_test_scores),   np.min(fake_scores)],
    "max_score":  [np.max(real_train_scores),  np.max(real_val_scores),
                   np.max(real_test_scores),   np.max(fake_scores)],
})

print("\nFinal summary table:")
print(summary_table.to_string(index=False))

summary_table.to_csv("/content/cgan_summary_table.csv", index=False)


# 24) Save training curve values
pd.DataFrame({
    "epoch": range(1, EPOCHS + 1),
    "disc_loss": disc_losses,
    "gen_loss": gen_losses,
    "d_real_val": val_real_hist,
    "d_fake": val_fake_hist
}).to_csv("/content/cgan_training_curves.csv", index=False)


# 25) Final saved outputs
print(
    "\nSaved: physiological_scores_results.csv, "
    "cgan_training_curves.png, score_distributions.png, cgan_summary_table.csv"
)

# Save all main preprocessing stages to CSV
real_df.to_csv("real_merged_cleaned.csv", index=False)
fake_df.to_csv("fake_merged_cleaned.csv", index=False)

real_model.to_csv("real_model_ready.csv", index=False)
fake_model.to_csv("fake_model_ready.csv", index=False)

real_scaled.to_csv("real_scaled.csv", index=False)
fake_scaled.to_csv("fake_scaled.csv", index=False)
```

```python
from sklearn.metrics import (
    roc_auc_score, roc_curve, precision_recall_curve,
    average_precision_score, confusion_matrix
)
from scipy.stats import ks_2samp

# Combined evaluation metrics
# real = 1, fake = 0
y_true   = np.concatenate([np.ones(len(real_test_scores)), np.zeros(len(fake_scores))])
y_scores = np.concatenate([real_test_scores, fake_scores])

# AUC-ROC:
# measures how well the scores separate real from fake across all possible thresholds
# closer to 1 = stronger separation, 0.5 = random guessing
auc_roc = roc_auc_score(y_true, y_scores)

# AUC-PR:
# area under the precision-recall curve
# useful when class balance matters; higher = better separation of positives (real cases)
auc_pr  = average_precision_score(y_true, y_scores)

# KS statistic:
# measures the maximum distance between the real-score distribution and fake-score distribution
# higher = stronger distributional separation
ks_stat, ks_p = ks_2samp(real_test_scores, fake_scores)

# Mean separation gap:
# simple difference between average real-test score and average fake score
# larger gap = clearer practical separation
separation_gap = np.mean(real_test_scores) - np.mean(fake_scores)

# Threshold-based classification using 10th percentile of real test
# this sets a cutoff based on the lower tail of the real distribution
threshold = np.percentile(real_test_scores, 10)

# convert continuous plausibility scores into class predictions
# score >= threshold -> predict real / plausible
# score <  threshold -> predict fake / implausible
y_pred = (y_scores >= threshold).astype(int)

# Confusion matrix:
# summarises classification outcomes at the chosen threshold
# [[TN, FP],
#  [FN, TP]]
cm = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm.ravel()

# Precision:
# of all samples predicted as real, how many were actually real
# high precision = few fake samples wrongly kept
precision = tp / (tp + fp)

# Recall / Sensitivity:
# of all real samples, how many were correctly kept as real
# high recall = few real samples wrongly rejected
recall = tp / (tp + fn)

# Specificity:
# of all fake samples, how many were correctly rejected
# high specificity = strong fake/OOD rejection
specificity = tn / (tn + fp)

# F1 score:
# harmonic mean of precision and recall
# useful when you want one balanced threshold-based score
f1 = 2 * precision * recall / (precision + recall)

# False positive rate:
# of all fake samples, how many were wrongly kept as real
# lower is better
fpr_val = fp / (fp + tn)

# Accuracy:
# overall proportion of correctly classified samples at the chosen threshold
# can be useful, but depends on the threshold choice
accuracy = (tp + tn) / (tp + tn + fp + fn)

# Print metrics
print("=" * 50)
print("FULL EVALUATION METRICS")
print("=" * 50)

print(f"\nDistribution-level metrics:")
print(f"  AUC-ROC              : {auc_roc:.4f}")
print(f"  AUC-PR               : {auc_pr:.4f}")
print(f"  KS statistic         : {ks_stat:.4f}  (p={ks_p:.2e})")
print(f"  Mean separation gap  : {separation_gap:.4f}")

print(f"\nThreshold-based metrics (threshold = 10th pct = {threshold:.4f}):")
print(f"  Sensitivity (recall) : {recall:.4f}   — % real correctly kept")
print(f"  Specificity          : {specificity:.4f}   — % fake correctly rejected")
print(f"  Precision            : {precision:.4f}")
print(f"  F1 score             : {f1:.4f}")
print(f"  False positive rate  : {fpr_val:.4f}   — % fake wrongly kept")
print(f"  Accuracy             : {accuracy:.4f}   — % of all samples correctly classified")

print(f"\nConfusion matrix (threshold = {threshold:.4f}):")
print(f"  True  Positives (real kept)    : {tp}")
print(f"  False Negatives (real rejected): {fn}")
print(f"  True  Negatives (fake rejected): {tn}")
print(f"  False Positives (fake kept)    : {fp}")

# =========================================================
# Plots
# =========================================================
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Model evaluation metrics", fontsize=13, fontweight="bold")

# ROC curve:
# plots true positive rate vs false positive rate across thresholds
# curve closer to top-left = better
fpr_c, tpr_c, _ = roc_curve(y_true, y_scores)
axes[0].plot(fpr_c, tpr_c, color='#185FA5', linewidth=2, label=f'AUC = {auc_roc:.4f}')
axes[0].plot([0, 1], [0, 1], color='gray', linestyle='--', alpha=0.5, label='Random classifier')
axes[0].set_xlabel('False positive rate')
axes[0].set_ylabel('True positive rate')
axes[0].set_title('ROC curve')
axes[0].legend()
axes[0].grid(alpha=0.3)
axes[0].spines[['top', 'right']].set_visible(False)

# Precision-Recall curve:
# shows trade-off between precision and recall across thresholds
# especially useful when positives are rarer or when false positives matter
prec_c, rec_c, _ = precision_recall_curve(y_true, y_scores)
axes[1].plot(rec_c, prec_c, color='#1D9E75', linewidth=2, label=f'AUC-PR = {auc_pr:.4f}')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-recall curve')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].spines[['top', 'right']].set_visible(False)

# Score distributions with threshold:
# visual comparison of real vs fake plausibility score distributions
# dashed line shows the chosen classification / OOD threshold
axes[2].hist(real_test_scores, bins=30, alpha=0.65, label='Real test',
             color='#1D9E75', edgecolor='white', density=True)
axes[2].hist(fake_scores, bins=30, alpha=0.65, label='Fake',
             color='#EF9F27', edgecolor='white', density=True)
axes[2].axvline(threshold, color='#A32D2D', linestyle='--',
                linewidth=1.5, label=f'Threshold ({threshold:.2f})')
axes[2].set_xlabel('Plausibility score')
axes[2].set_ylabel('Density')
axes[2].set_title('Score distributions')
axes[2].legend()
axes[2].grid(alpha=0.3)
axes[2].spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.savefig('/content/full_evaluation.png', dpi=150, bbox_inches='tight')
plt.show()

# Save metrics to CSV for reporting / appendix use
metrics_df = pd.DataFrame([{
    'auc_roc': auc_roc,
    'auc_pr': auc_pr,
    'ks_statistic': ks_stat,
    'ks_p_value': ks_p,
    'separation_gap': separation_gap,
    'threshold_10pct': threshold,
    'sensitivity': recall,
    'specificity': specificity,
    'precision': precision,
    'f1_score': f1,
    'false_positive_rate': fpr_val,
    'accuracy': accuracy,
    'tp': tp,
    'fn': fn,
    'tn': tn,
    'fp': fp
}])

metrics_df.to_csv('/content/evaluation_metrics.csv', index=False)
print("\nSaved: evaluation_metrics.csv, full_evaluation.png")
```

```python
# Seed Stability Testing
# Trains the full model (warm-up + adversarial) across multiple random seeds
# and reports mean ± std across runs to confirm reproducibility.

import gc

SEEDS        = [42, 7, 123, 99, 2024]
seed_results = []

print("=" * 60)
print("SEED STABILITY TESTING")
print(f"Running {len(SEEDS)} independent seeds: {SEEDS}")
print("=" * 60)

for seed_idx, seed in enumerate(SEEDS):
    print(f"\n── Seed {seed} ({seed_idx + 1}/{len(SEEDS)}) ──")

    # Set all random seeds
    np.random.seed(seed)
    tf.random.set_seed(seed)

    #  Rebuild models fresh for each seed
    gen_s  = build_generator(noise_dim, cond_dim)
    disc_s = build_discriminator(cond_dim)

    bce_s = tf.keras.losses.BinaryCrossentropy()
    mse_s = tf.keras.losses.MeanSquaredError()

    # Unsupervised warm-up
    warmup_opt_s = tf.keras.optimizers.Adam(0.001)
    warmup_ds_s  = (tf.data.Dataset.from_tensor_slices(train_array)
                    .shuffle(1000, seed=seed).batch(64))

    for epoch in range(80):
        for real_batch in warmup_ds_s:
            real_ffr_b  = real_batch[:, 0:1]
            real_cond_b = real_batch[:, 1:]

            idx_ffr      = tf.random.shuffle(tf.range(tf.shape(real_ffr_b)[0]))
            shuffled_ffr = tf.gather(real_ffr_b, idx_ffr)

            idx_cond      = tf.random.shuffle(tf.range(tf.shape(real_cond_b)[0]))
            shuffled_cond = tf.gather(real_cond_b, idx_cond)

            with tf.GradientTape() as tape:
                real_sc  = disc_s([real_ffr_b,  real_cond_b],   training=True)
                shuf_f   = disc_s([shuffled_ffr, real_cond_b],  training=True)
                shuf_c   = disc_s([real_ffr_b,  shuffled_cond], training=True)
                loss = (bce_s(tf.ones_like(real_sc)  * 0.95, real_sc)  +
                        bce_s(tf.zeros_like(shuf_f)  + 0.05, shuf_f)   +
                        bce_s(tf.zeros_like(shuf_c)  + 0.05, shuf_c)   +
                        mse_s(tf.ones_like(real_sc), real_sc) * 0.5) / 3.0
            grads = tape.gradient(loss, disc_s.trainable_variables)
            warmup_opt_s.apply_gradients(zip(grads, disc_s.trainable_variables))

    wu_real = float(tf.reduce_mean(disc_s(
        [train_array[:, 0:1], train_array[:, 1:]], training=False)))
    print(f"  After warm-up — D(real): {wu_real:.3f}")

    # Adversarial training
    gen_opt_s  = tf.keras.optimizers.Adam(learning_rate=2e-4, beta_1=0.5)
    disc_opt_s = tf.keras.optimizers.Adam(learning_rate=8e-5, beta_1=0.5)

    train_ds_s = (tf.data.Dataset.from_tensor_slices(train_array)
                  .shuffle(1000, seed=seed).batch(32))

    @tf.function
    def disc_step_s(real_batch):
        batch_size  = tf.shape(real_batch)[0]
        real_ffr_b  = real_batch[:, 0:1]
        real_cond_b = real_batch[:, 1:]
        noise       = tf.random.normal([batch_size, noise_dim])
        gen_ffr     = gen_s([noise, real_cond_b], training=False)
        idx         = tf.random.shuffle(tf.range(batch_size))
        shuf_ffr    = tf.gather(real_ffr_b, idx)
        n = lambda t: tf.clip_by_value(
            t + tf.random.normal(tf.shape(t), stddev=0.003), 0, 1)
        with tf.GradientTape() as tape:
            rs  = disc_s([n(real_ffr_b), n(real_cond_b)], training=True)
            gs  = disc_s([n(gen_ffr),    n(real_cond_b)], training=True)
            ss  = disc_s([n(shuf_ffr),   n(real_cond_b)], training=True)
            adv = (bce_s(tf.ones_like(rs)  * 0.95, rs) +
                   bce_s(tf.zeros_like(gs) + 0.05, gs) +
                   bce_s(tf.zeros_like(ss) + 0.05, ss)) / 3.0
            pen = tf.reduce_mean(tf.square(tf.maximum(0.0, 0.95 - rs)))
            loss = adv + 0.4 * pen
        grads = tape.gradient(loss, disc_s.trainable_variables)
        disc_opt_s.apply_gradients(zip(grads, disc_s.trainable_variables))
        return loss

    @tf.function
    def gen_step_s(real_batch):
        batch_size  = tf.shape(real_batch)[0]
        real_cond_b = real_batch[:, 1:]
        noise       = tf.random.normal([batch_size, noise_dim])
        with tf.GradientTape() as tape:
            gen_ffr = gen_s([noise, real_cond_b], training=True)
            gs      = disc_s([gen_ffr, real_cond_b], training=False)
            loss    = bce_s(tf.ones_like(gs) * 0.95, gs)
        grads = tape.gradient(loss, gen_s.trainable_variables)
        gen_opt_s.apply_gradients(zip(grads, gen_s.trainable_variables))
        return loss

    for epoch in range(400):
        for real_batch in train_ds_s:
            for _ in range(2):
                disc_step_s(real_batch)
            gen_step_s(real_batch)

        if (epoch + 1) % 100 == 0:
            v_real = float(tf.reduce_mean(disc_s(
                [val_array[:, 0:1], val_array[:, 1:]], training=False)))
            v_fake = float(tf.reduce_mean(disc_s(
                [fake_array[:, 0:1], fake_array[:, 1:]], training=False)))
            print(f"  Epoch {epoch+1}/400 | D(val_real): {v_real:.3f} | D(fake): {v_fake:.3f}")

    # Score and evaluate
    def get_scores_s(arr):
        ffr  = np.clip(arr[:, 0:1].astype("float32"), 0, 1)
        cond = np.clip(arr[:, 1:].astype("float32"),  0, 1)
        return disc_s.predict([ffr, cond], verbose=0).flatten()

    real_test_sc = get_scores_s(test_array)
    fake_sc      = get_scores_s(fake_array)

    from sklearn.metrics import roc_auc_score, average_precision_score
    from scipy.stats import ks_2samp

    auc_roc  = roc_auc_score(
        np.concatenate([np.ones(len(real_test_sc)), np.zeros(len(fake_sc))]),
        np.concatenate([real_test_sc, fake_sc])
    )
    auc_pr   = average_precision_score(
        np.concatenate([np.ones(len(real_test_sc)), np.zeros(len(fake_sc))]),
        np.concatenate([real_test_sc, fake_sc])
    )
    ks_stat, _ = ks_2samp(real_test_sc, fake_sc)
    threshold  = np.percentile(real_test_sc, 10)
    fake_det   = np.mean(fake_sc < threshold)
    gap        = np.mean(real_test_sc) - np.mean(fake_sc)

    run = {
        "seed":          seed,
        "real_mean":     float(np.mean(real_test_sc)),
        "fake_mean":     float(np.mean(fake_sc)),
        "gap":           float(gap),
        "auc_roc":       float(auc_roc),
        "auc_pr":        float(auc_pr),
        "ks_stat":       float(ks_stat),
        "fake_det_10th": float(fake_det),
    }
    seed_results.append(run)

    print(f"  Real mean: {run['real_mean']:.4f} | Fake mean: {run['fake_mean']:.4f} | "
          f"Gap: {run['gap']:.4f} | AUC-ROC: {run['auc_roc']:.4f} | "
          f"Fake det: {run['fake_det_10th']*100:.1f}%")

    # Free memory before next seed
    del gen_s, disc_s
    gc.collect()
    tf.keras.backend.clear_session()

# Summary table
print("\n" + "=" * 60)
print("SEED STABILITY SUMMARY")
print("=" * 60)

metrics = ["real_mean", "fake_mean", "gap", "auc_roc", "auc_pr", "ks_stat", "fake_det_10th"]
labels  = ["Real mean", "Fake mean", "Gap", "AUC-ROC", "AUC-PR", "KS Stat", "Fake det (10th)"]

results_df = pd.DataFrame(seed_results)
print(f"\n{'Seed':<8}", end="")
for m in metrics:
    print(f"{m:<18}", end="")
print()
print("-" * (8 + 18 * len(metrics)))

for row in seed_results:
    print(f"{row['seed']:<8}", end="")
    for m in metrics:
        val = row[m]
        print(f"{val:<18.4f}", end="")
    print()

print("-" * (8 + 18 * len(metrics)))
print(f"{'Mean':<8}", end="")
for m in metrics:
    print(f"{np.mean([r[m] for r in seed_results]):<18.4f}", end="")
print()
print(f"{'Std':<8}", end="")
for m in metrics:
    print(f"{np.std([r[m] for r in seed_results]):<18.4f}", end="")
print()

# Save
results_df.to_csv("/content/seed_stability_results.csv", index=False)
print("\nSaved: seed_stability_results.csv")
```
